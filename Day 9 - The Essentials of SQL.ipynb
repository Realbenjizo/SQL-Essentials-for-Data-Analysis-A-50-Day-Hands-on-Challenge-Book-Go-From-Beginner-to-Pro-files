## Question 1: **Answer**

```python
import sqlite3
```

```python
def add_data_to_database():
    try:
        # Connect to the database
        with sqlite3.connect('sales_team.db') as conn:
            cursor = conn.cursor()

           # Create the table
            cursor.execute('''
            CREATE TABLE IF NOT EXISTS sales (
                id INTEGER PRIMARY KEY,
                name TEXT,
                seller_id INTEGER,
                sales_value REAL
            )
            ''' )
            # Insert data into the table
            cursor.executemany('''
            INSERT INTO sales (id, name, seller_id, sales_value) VALUES (?, ?, ?, ?)
            ''', [
                (1, 'John', 101, 20000),
                (2, 'Mary', 102, 50000),
                (3, 'Jerry', 103, 3000),
                (4, 'Jerry', 103, 6000),
                (5, 'John', 101, 15000)
                ]
                )
            conn.commit()
            print("Table created successfully.")
    except Exception as e:
        print(f"Error: {e}")
    finally:
        conn.close()

# Calling the function
add_data_to_database()
```

## Question 2: **Answer**

```python
%%sql
SELECT
    seller_id,
    name,
    CAST(SUM(sales_value) AS INTEGER) AS total_sales
FROM sales
GROUP BY seller_id, name
ORDER BY total_sales DESC;
```

## Question 3: **Answer**

```python
%%sql
SELECT
    seller_id,
    name,
    AVG(sales_value) AS average_sales
FROM sales
GROUP BY seller_id, name;
```

## Question 4: **Answer**

```python
%%sql
SELECT
    seller_id,
    name,
    SUM(sales_value) AS total_sales
FROM  sales
GROUP BY  seller_id, name
HAVING  SUM(sales_value) >= 50000;
```

## Question 5: **Answer**

```python
%%sql
SELECT
    seller_id,
    name,
    SUM(sales_value) AS total_sales
FROM  sales
WHERE  name = 'Jerry'
GROUP BY  seller_id, name
HAVING  SUM(sales_value) > 5000;
```
