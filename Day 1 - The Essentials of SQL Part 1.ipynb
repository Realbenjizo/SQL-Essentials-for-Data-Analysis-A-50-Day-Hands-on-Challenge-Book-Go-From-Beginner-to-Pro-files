```python
import sqlite3
```

```python
# Create a SQLite database if it doesn't exist
connection = sqlite3.connect('employee_records.db')
```

```python
# Create a cursor object
cursor = connection.cursor()
```

```python
cursor.execute('''
                  CREATE TABLE IF NOT EXISTS employees (
                  id INTEGER PRIMARY KEY,
                  name TEXT  NOT NULL,
                  position TEXT,
                  salary REAL)

                  '''
                  )
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS employees (
        id INTEGER PRIMARY KEY,  -- Ensures a unique identifier
        name TEXT NOT NULL CHECK(typeof(name) = 'text' AND name GLOB '*[^0-9]*'),
        position TEXT NOT NULL CHECK(typeof(position) = 'text' AND position GLOB '*[^0-9]*'),
        salary REAL CHECK(salary > 0)  -- Salary must be a positive number
    ) STRICT;
''')
```

```python
# Creating an employees list
data = [
    (101, 'Rai', 'Developer', 75000),
    (102, 'Bobby', 'Manager', 85000),
    (103, 'Charles', 'Analyst', 65000)
]

# Inserting data into the table
cursor.executemany('INSERT INTO employees VALUES (?, ?, ?, ?)', data)

# Committing changes to the database
connection.commit()
```

```python
cursor.execute('SELECT * FROM employees')

# Returns a single row
cursor.fetchone()
```

```python
cursor.execute('SELECT * FROM employees')

# Returns all the rows
cursor.fetchall()
```

```python
cursor.execute('SELECT * FROM employees')

# Returning two rows
cursor.fetchmany(size=2)
```
