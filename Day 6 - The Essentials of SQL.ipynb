## Question 1: **Answer**

```python
import sqlite3
```

```python
def add_data_to_database():
    try:
        # Connect to the SQLite database
        with sqlite3.connect('sales_team_data.db') as conn:
            cursor = conn.cursor()
            # Create the table
            cursor.execute('''
            CREATE TABLE IF NOT EXISTS sales (
                id INTEGER PRIMARY KEY,
                sales_person TEXT,
                age INTEGER,
                year INTEGER,
                sales_value INTEGER
            )
            ''')
            # Insert data into the table
            cursor.executemany('''
            INSERT INTO sales (id, sales_person, age, year, sales_value)
            VALUES (?, ?, ?, ?, ?)
            ''', [
                (1, 'John', 28, 2001, 1500),
                (2, 'Mary', 34, 2023, 2200),
                (3, 'Ken', 29, 2022, 1800),
                (4, 'John', 28, 2021, 2500),
                (5, 'Sarah', 31, 2023, 1300),
                (6, 'Mary', 34, 2022, 2600),
                (7, 'Ken', 29, 2002, 1400),
                (8, 'Sarah', 31, 2001, 1900)
            ])
            conn.commit()
            print("Table created successfully.")
    except Exception as e:
        print(f"Error: {e}")
    finally:
        conn.close()

# Calling the function
add_data_to_database()
```

## 1. Question 2: **Answer**
Rewrite the query using the IN operator.

```python
%%sql
SELECT sales_value
FROM sales
WHERE sales_person = 'John'
      OR sales_person = 'Mary'
      OR sales_person = 'Ken';
```

```python
%%sql
SELECT sales_value
FROM sales
WHERE sales_person IN ('John', 'Mary', 'Ken');
```

## Question 3: **Answer**

```python
%%sql
SELECT
    sales_person,
    sales_value
FROM sales
WHERE sales_value BETWEEN 1400 AND 2500
ORDER BY sales_value DESC;
```

## Question 4: **Answer**

```python
%%sql
SELECT
    sales_person,
    sales_value
FROM sales
WHERE sales_person IN ('John', 'Mary', 'Ken')
      AND sales_value BETWEEN 1400 AND 2500
ORDER BY sales_value DESC;
```
