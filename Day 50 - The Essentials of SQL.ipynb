```python
%%sql
WITH count_duplicates AS (
    SELECT
        COUNT(*) AS number_of_duplicates
    FROM insurance
    GROUP BY ID
    HAVING COUNT(*) > 1

)
SELECT
    SUM(number_of_duplicates) AS total_duplicates
FROM count_duplicates;
```

```python
%%sql
-- Creating partitions for each ID
WITH ranked_rows AS (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY ID ORDER BY ROWID) AS ranked
    FROM insurance
)
-- Deleting all duplicates ranked above 1
DELETE FROM insurance
WHERE ROWID IN (
    SELECT ROWID
    FROM ranked_rows
    WHERE ranked > 1
);
```

## Question 2 Part B: **Answer**

```python
%%sql
WITH date_parts AS (
    SELECT
        ID,
        SUBSTR(BIRTH, 1, 2) AS day,
        CASE SUBSTR(BIRTH, 3, 3)
            WHEN 'JAN' THEN '01'
            WHEN 'FEB' THEN '02'
            WHEN 'MAR' THEN '03'
            WHEN 'APR' THEN '04'
            WHEN 'MAY' THEN '05'
            WHEN 'JUN' THEN '06'
            WHEN 'JUL' THEN '07'
            WHEN 'AUG' THEN '08'
            WHEN 'SEP' THEN '09'
            WHEN 'OCT' THEN '10'
            WHEN 'NOV' THEN '11'
            WHEN 'DEC' THEN '12'
        END AS month,
        SUBSTR(BIRTH, 6, 2) AS year
    FROM insurance
),
full_year AS (
    SELECT
        id,
        day,
        month,
        CASE WHEN year >= '00' THEN '19' || year
             ELSE '20' || year
        END AS full_year
    FROM date_parts
)
-- Updating the birth column
UPDATE insurance
SET BIRTH = DATE((SELECT full_year||'-'||month||'-'||day
            FROM full_year
            WHERE insurance.id = full_year.id));
```

## Question 5: **Answer**

```python
%%sql
-- Categorize data into two groups
WITH drivers_and_kids AS (
    SELECT
        CASE
            WHEN KIDSDRIV > 0 THEN 'With Kid Driver'
            ELSE 'Without Kid Driver'
        END AS kid_driver_status,
        COUNT(*) AS total_claims_count,
        SUM(CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL))
        AS total_claim_amount
    FROM insurance
    WHERE CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL) > 0
    GROUP BY kid_driver_status

),
-- Aggregating total claims count and claims amount from the two categories
total_claims AS (
    SELECT
        SUM(total_claims_count) AS overall_claims_count,
        SUM(total_claim_amount) AS overall_claims_amount
    FROM drivers_and_kids
)
-- Returning totals and percentage share of each group
SELECT
    d.kid_driver_status,
    d.total_claims_count,
    ROUND((d.total_claims_count * 100.0) / t.overall_claims_count, 2)
    AS percentage_of_total_claims_count,
    d.total_claim_amount,
    ROUND((d.total_claim_amount * 100.0) /overall_claims_amount , 2)
    AS percentage_of_total_claim_amount
FROM drivers_and_kids d
CROSS JOIN total_claims t
ORDER BY percentage_of_total_claims_count DESC;
```

## Question 6: **Answer**

```python
%%sql
-- Cleaning the CLM_AMT column
WITH cleaned_data AS (
    SELECT
        CAR_TYPE,
        CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL)
        AS cleaned_clm_amt
    FROM insurance
    WHERE CLAIM_FLAG > 0
),

filtered_data AS (
    SELECT CAR_TYPE, cleaned_clm_amt
    FROM cleaned_data
    WHERE cleaned_clm_amt > 0
),
flagged_claims AS (
    SELECT
        CAR_TYPE,
        COUNT(*) AS total_flagged_claims_count,
        SUM(cleaned_clm_amt) AS claims_amounts
    FROM filtered_data
    GROUP BY CAR_TYPE
),
-- Total count of all flagged types
total_flagged AS (
    SELECT SUM(total_flagged_claims_count) AS overall_flagged_claims_count
    FROM flagged_claims
)
-- Returning counts and calculating percentage share of each car type
SELECT
    f.CAR_TYPE,
    f.total_flagged_claims_count,
    ROUND((f.total_flagged_claims_count * 100.0) / t.overall_flagged_claims_count, 2)
        AS percentage_of_total_flagged_claims,
    f.claims_amounts
FROM flagged_claims f
CROSS JOIN total_flagged t
WHERE
    f.total_flagged_claims_count = (SELECT MAX(total_flagged_claims_count)
                                    FROM flagged_claims)
    OR
    f.total_flagged_claims_count = (SELECT MIN(total_flagged_claims_count)
                                    FROM flagged_claims)
ORDER BY f.total_flagged_claims_count DESC;
```

## Question 7: **Answer**

```python
%%sql
-- calculating average claim
WITH avg_claims AS (
    SELECT
        CAR_TYPE,
        OCCUPATION,
        ROUND(AVG(CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL)), 2)
        AS AVG_CLAIM_AMOUNT
    FROM insurance
    WHERE CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL) > 0
    GROUP BY CAR_TYPE, OCCUPATION
),
-- raking car types per average claim amount
ranked_claims AS (
    SELECT
        CAR_TYPE,
        OCCUPATION,
        avg_claim_amount,
        RANK() OVER (PARTITION BY CAR_TYPE ORDER BY avg_claim_amount DESC)
        AS car_rank
    FROM avg_claims
)
-- Selecting car types number with the highest average claim amount
SELECT
    CAR_TYPE,
    OCCUPATION,
    avg_claim_amount
FROM ranked_claims
WHERE car_rank = 1
ORDER BY avg_claim_amount DESC;
```

```python
%%sql
WITH valid_car_owners AS (
    -- Filter individuals whose car value is at least 30% of their home value
    SELECT
        GENDER,
        HOME_VAL,
        CAST(REPLACE(REPLACE(BLUEBOOK, '$', ''), ',', '') AS REAL) AS car_value,
        CAST(REPLACE(REPLACE(HOME_VAL, '$', ''),',', '') AS REAL) AS home_value,
        CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL) AS claim_amount
    FROM insurance
    WHERE home_value >= 0.3 * car_value

)
-- Return total claims attributted to male or female clients
SELECT
    SUM(CASE WHEN GENDER = 'F' THEN claim_amount  ELSE 0 END)
    AS total_claims_female,
    SUM(CASE WHEN GENDER = 'M' THEN claim_amount ELSE 0 END)
    AS total_claims_male
FROM valid_car_owners
WHERE claim_amount > 0;
```

## Question 9: **Answer**

```python
%%sql
-- Aggregating current and old claims and grouping by private and commercial
WITH educated_drivers_claim AS (
    SELECT
        CAR_USE,
        SUM(CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL))
        AS total_claims,
        SUM(CAST(REPLACE(REPLACE(OLDCLAIM, '$', ''), ',', '') AS REAL))
        AS total_old_claims
    FROM insurance
    WHERE EDUCATION IN ('Bachelors', 'Masters')
    AND CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL) > 0
    GROUP BY CAR_USE
),
--  Calculating total current and old claims
total_claims AS (
    SELECT
           SUM(CAST(REPLACE(REPLACE(CLM_AMT, '$', ''), ',', '') AS REAL))
           AS overall_claims,
           SUM(CAST(REPLACE(REPLACE(OLDCLAIM, '$', ''), ',', '') AS REAL))
           AS overall_old_claims
    FROM insurance
)
-- Returning claims share per car use category
SELECT
    e.CAR_USE,
    e.total_claims,
    ROUND((e.total_claims * 100.0) / t.overall_claims, 2)
    AS percentage_of_overall_current_claims,
    ROUND((e.total_old_claims * 100.0) / t.overall_old_claims, 2)
    AS percentage_share_of_overall_old_claims
FROM educated_drivers_claim e
CROSS JOIN total_claims t
ORDER BY e.total_claims DESC;
```

## Questions 10: **Answer**

```python
%%sql
 -- Assigns a unique row number to each row in ascending order of ID
WITH numbered_claims AS (
    SELECT
        ROW_NUMBER() OVER (ORDER BY ID) AS row_number,
        CLAIM_FLAG
    FROM insurance
),
-- Identifying consecutive zero-flagged claims and assigns them a group number
consecutive_groups AS (
    SELECT
        CLAIM_FLAG,
        row_number - ROW_NUMBER() OVER (PARTITION BY CLAIM_FLAG ORDER BY row_number)
        AS group_number
    FROM numbered_claims
    WHERE CLAIM_FLAG = 0
),
-- Counting the length of each consecutive zero-flagged group
grouped_data AS (
    SELECT
        COUNT(*) AS longest_consecutive_zero_flagged
    FROM consecutive_groups
    GROUP BY group_number
)
SELECT
    Max(longest_consecutive_zero_flagged) AS longest_flagged_zeros
FROM grouped_data;
```
