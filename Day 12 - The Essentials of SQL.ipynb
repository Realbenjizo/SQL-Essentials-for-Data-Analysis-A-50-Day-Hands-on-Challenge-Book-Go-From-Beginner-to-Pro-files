## Question 1: **Answer**

```python
import sqlite3
import csv
```

```python
def load_employees_data(data):
    try:
        # Connecting to database
        conn = sqlite3.connect("company_records.db")
        cursor = conn.cursor()
        # Creating table
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS employees (
                employeeID INTEGER PRIMARY KEY,
                first_name TEXT,
                last_name TEXT,
                email TEXT,
                years_worked INTEGER,
                flagged_for_layoff INTEGER
            )
        """)

        with open(data, "r") as csv_file:
            csv_reader = csv.reader(csv_file)
            # Skipping the header row
            next(csv_reader)
            for row in csv_reader:
                cursor.execute("INSERT INTO employees VALUES (?, ?, ?, ?, ?, ?)", row)
        conn.commit()
    except Exception as e:
        print(f"Error: {e}")
    finally:
        conn.close()

# Loading data
data = "/content/company_employees_data.csv"
load_employees_data(data)
```

## Question 2: **Answer**

```python
%%sql
SELECT
    First_Name,
    Last_Name,
    Email
FROM employees
WHERE First_Name IN ('Robert','Benjamin');
```

## Question 3 **Answer**

```python
%%sql
SELECT
    count(first_name) AS over_five_years_employees
FROM employees
WHERE years_worked >= 5;
```

## Question 4: **Answer**

```python
%%sql
SELECT
    first_name,
    last_name,
    years_worked,
    flagged_for_layoff
FROM employees
WHERE  first_Name = 'Grace' AND last_name = 'Hughes';
```

## Question 5: **Answer**

```python
%%sql
SELECT
    first_name,
    last_name,
    years_worked,
    flagged_for_layoff
FROM employees
WHERE years_worked > 10 AND flagged_for_layoff = "No";
```

## Question 6: **Answer**

```python
%%sql
SELECT first_name, last_name, email, years_worked
FROM employees
WHERE years_worked = (SELECT MAX(years_worked) FROM employees);
```
